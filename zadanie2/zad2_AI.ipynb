```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

print(tf.__version__)
#%matplotlib inline

training_data = np.load('TF_dataset.npz')
input_size = 2
output_size = 1

model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation='sigmoid'),  # ukryta warstwa
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(1)                         # wyjście – bez aktywacji
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),  # zmiana optymalizatora
              loss=tf.keras.losses.MeanAbsoluteError())  # zmiana funkcji straty
model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)

model.layers[0].get_weights()

plt.plot(np.squeeze(model.predict_on_batch(training_data['inputs'])), np.squeeze(training_data['targets']), color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
