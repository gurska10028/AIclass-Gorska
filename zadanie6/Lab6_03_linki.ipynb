# Notebook 1: Zbieranie treści i istotnych linków ze strony

```python
import requests
from bs4 import BeautifulSoup
import ollama
from urllib.parse import urljoin
import json
```

```python
def get_page_text_and_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
    links = [urljoin(url, a.get('href')) for a in soup.find_all('a', href=True)]
    return text, links
```

```python
system_prompt_links = """
Zadanie: wybierz najważniejsze linki dla prezentacji firmy (np. o nas, misja, produkty).
Zwróć wynik jako czysty JSON:
{ \"ważne_linki\": [\"https://example.com/about\", \"https://example.com/products\"] }
"""
```

```python
def filter_links_with_model(links, model='llama3'):
    prompt = json.dumps(links[:20])
    messages = [
        {"role": "system", "content": system_prompt_links},
        {"role": "user", "content": f"Wybierz z listy ważne linki: {prompt}"}
    ]
    response = ollama.chat(model=model, messages=messages)
    return json.loads(response['message']['content'])
```
